#  Visualisasi

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(17,10))

plt.subplot(2,2,1)
sns.barplot(x='prov', y='fourty', data=data.groupby(data['prov']).sum().fourty.reset_index(), palette='Purples_r')
plt.ylabel('Total Data')
plt.xlabel('Provinsi')
plt.xticks(rotation=80)
plt.title('Mama Tau Sendiri')
```
